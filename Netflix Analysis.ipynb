```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'])
```

```python
df.head(2)
```

```python
df['Release_Date']=df['Release_Date'].dt.year
```

```python
df.head(2)
```

```python
df.drop(['Overview','Original_Language','Poster_Url'], axis =1, inplace=True)
```

```python
df.columns
```

```python
df.head(10).to_csv('First_10_rows.csv', index=False)
```

```python
bins = [0, 4, 6, 7.5, 10]
labels = ["Not_popular", "Below_Average", "Average", "Famous"]

df["Vote_Average"] = pd.cut(df["Vote_Average"], bins=bins, labels=labels, right=False)
```

```python
df.head(2)
```

```python
df.head(5)
```

```python
df.head(2)
```

```python
df['Vote_Average'].value_counts()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
# Split into list
df["Genre"] = df["Genre"].str.split(", ")

# Explode into multiple rows
df = df.explode("Genre")
```

```python
df.head(5)
```

```python
df = df.reset_index(drop=True) #Reset the index
```

```python
df.head(5)
```

```python
#casting column into category 
df['Genre'] = df['Genre'].astype('category')
```

```python
df.info()
```

```python
df.nunique()
```

```python
df.info()
```

```python
df.to_csv('cleaned_mymoviedb.csv', index=False)
```

```python
year_counts = df["Release_Date"].value_counts()

print("Movies per year:")
print(year_counts)

# get the year(s) with highest number of movies
max_count = year_counts.max()
top_years = year_counts[year_counts == max_count]

print("\nYear(s) with highest number of movies:")
print(top_years)
```
